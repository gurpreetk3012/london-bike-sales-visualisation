#### Mapping Numerical Codes to Descriptive Labels

To enhance the interpretability of the dataset, numerical codes for categorical variables such as `season`, `weather`, `is_holiday`, and `is_weekend` were mapped to descriptive labels. This was achieved using Python dictionaries and the pandas `map()` function.

```python
# Mapping numerical codes to descriptive labels for better interpretability
season_dict = {
    '0.0': 'Spring',
    '1.0': 'Summer',
    '2.0': 'Autumn',
    '3.0': 'Winter'
}

weather_dict = {
    '1.0': 'Clear',
    '2.0': 'Partly_cloudy',
    '3.0': 'Cloudy',
    '4.0': 'Light_rain',
    '7.0': 'Rain',
    '10.0': 'Foggy',
    '26.0': 'Snow',
}

bikes.season = bikes.season.astype('str')
bikes.season = bikes.season.map(season_dict)

bikes.weather = bikes.weather.astype('str')
bikes.weather = bikes.weather.map(weather_dict)
```

```python
bikes.is_holiday.value_counts()
```

```python
bikes.is_weekend.value_counts()
```

```python
is_holiday_dict = {
    '0.0': 'No',
    '1.0': 'Yes'
}

is_weekend_dict = {
    '0.0': 'No',
    '1.0': 'Yes'
}

bikes.is_holiday = bikes.is_holiday.astype('str')
bikes.is_holiday = bikes.is_holiday.map(is_holiday_dict)

bikes.is_weekend = bikes.is_weekend.astype('str')
bikes.is_weekend = bikes.is_weekend.map(is_weekend_dict)
```
